```python
%%html

<!-- What is What? -->
<div>
    <h1>(William inspired) JavaScript Basics Blog</h1>
    <h2>What is What?</h2>
    <p>This section explains JSON objects, arrays, input/output, data types, variables, and operators in JavaScript.</p>
</div>

<!-- JSON Objects -->
<div>
    <h3>JSON Objects</h3>
    <p>JSON objects store data in key-value pairs.</p>
    <script>
        // Example JSON object
        let car = {
            color: "blue",
            yearMade: 2023,
            maker: "Tesla",
            isElectric: true,
            canFly: false
        };
        console.log("Car Object:", car); // Displays the car object
    </script>
</div>

<!-- JSON Arrays -->
<div>
    <h3>JSON Arrays</h3>
    <p>JSON arrays store multiple values in a list format.</p>
    <script>
        // Example JSON array
        (function() {
            let numbers = [15, 30, 45, 60];
            console.log("Second Number in Array:", numbers[1]); // Displays 30
        })();
    </script>
</div>

<!-- Input/Output -->
<div>
    <h3>Input/Output</h3>
    <p>Allows users to enter a value, which is then displayed in the console.</p>
    <label for="inputNumber">Enter a Number:</label>
    <input type="number" id="inputNumber" name="number" step="1" value="1">
    <button onclick="processInput()">Submit</button>
    <script>
        function processInput() {
            // Get the input value and display it in the console
            let userInput = document.getElementById("inputNumber").value;
            console.log("User entered:", userInput);
        }
    </script>
</div>

<!-- Data Types -->
<div>
    <h3>Data Types</h3>
    <p>JavaScript has different data types for different purposes.</p>
    <script>
        // Different data types
        let score = 100;                // Number
        const playerName = "Alex";      // String
        let isActive = true;            // Boolean
        let weapon;                     // Undefined
        let inventorySlot = null;       // Null

        console.log("Score:", score);
        console.log("Player Name:", playerName);
        console.log("Is Active:", isActive);
        console.log("Weapon:", weapon);
        console.log("Inventory Slot:", inventorySlot);
    </script>
</div>

<!-- Variables and Scope -->
<div>
    <h3>Variables and Scope</h3>
    <p>JavaScript uses <code>let</code>, <code>const</code>, and <code>var</code> to declare variables.</p>
    <script>
        if (typeof globalVar === 'undefined') {
            var globalVar = 42; // 'var' variables are accessible outside their block
        }

        function demoScope() {
            let localVar = 5;            // 'let' variables are block-scoped
            const constantVar = 3.14;    // 'const' variables are block-scoped and cannot change

            console.log("Inside function:");
            console.log("Global Var:", globalVar);
            console.log("Local Var:", localVar);
            console.log("Constant Var:", constantVar);
        }

        demoScope();
        console.log("Outside function, Global Var:", globalVar);
    </script>
</div>

<!-- Operators -->
<div>
    <h3>Operators</h3>
    <p>Operators allow for arithmetic and logic operations.</p>
    <script>
        function calculate(num1, num2) {
            console.log(`${num1} + ${num2} = ${num1 + num2}`); // Addition
            console.log(`${num1} - ${num2} = ${num1 - num2}`); // Subtraction
            console.log(`${num1} * ${num2} = ${num1 * num2}`); // Multiplication
            console.log(`${num1} / ${num2} = ${num1 / num2}`); // Division
            console.log(`${num1} % ${num2} = ${num1 % num2}`); // Modulus
        }

        calculate(12, 4); // Example calculation
    </script>
</div>
```
